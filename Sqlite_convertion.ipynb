```python
# Age ->
'''
import sqlite3, csv
conn = sqlite3.connect("main.db")
cur = conn.cursor()
sql = """
CREATE TABLE Age_chart (
    Age_low INTEGER,
    Age_high INTEGER,
    Male INTEGER,
    Female INTEGER
    )
"""
cur.execute(sql)
count = 0
with open("Age_chart.csv", "r") as file:
    for row in file:
        #print(row)
        count += 1
        if count == 1:
            continue
        #print(row)
        cur.execute("INSERT INTO Age_chart VALUES (?,?,?,?)", row.split(","))
        conn.commit()
conn.close()
'''
```
